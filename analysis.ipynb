```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression ,  Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
df.columns
```

```python
df.head(2)
```

```python
features = ['study_hours_per_day','attendance_percentage','mental_health_rating','sleep_hours','part_time_job']
```

```python
target = "exam_score"
```

```python
df_model = df[features + [target].copy()]
```

```python
df_model
```

```python
le = LabelEncoder()
```

```python
df_model["part_time_job"]=le.fit_transform(df_model["part_time_job"])
```

```python
df_model
```

```python
x = df_model[features]
```

```python
y = df_model[target]
```

```python
x_train, x_test , y_train , y_test = train_test_split(x,y, test_size=0.2)
```

```python
len(y_test)
```

```python
len(y_train)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
import joblib
models = {
    "LinearRegression": {
        "model": LinearRegression(),
        "params": {}
    },
    "DecisionTree": {
        "model": DecisionTreeRegressor(random_state=42),
        "params": {"max_depth": [3, 5, 10], "min_samples_split": [2, 5]}
    },
    "RandomForest": {
        "model": RandomForestRegressor(random_state=42),
        "params": {"n_estimators": [50, 100], "max_depth": [5, 10]}
    },
    "GradientBoosting": {
        "model": GradientBoostingRegressor(random_state=42),
        "params": {"n_estimators": [100], "max_depth": [3, 5], "learning_rate": [0.05, 0.1]}
    }
}

best_models = []
for name, config in models.items():
    print(f"Training {name}...")
    grid = GridSearchCV(config["model"], config["params"], cv=5, scoring="neg_mean_squared_error")
    grid.fit(x_train, y_train)
    y_pred = grid.predict(x_test)
    out_of_range = int(np.sum((y_pred < 0) | (y_pred > 100)))
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2   = r2_score(y_test, y_pred)
    best_models.append({
        "model": name,
        "estimator": grid.best_estimator_,
        "best_params": grid.best_params_,
        "rmse": rmse,
        "r2": r2,
        "out_of_range": out_of_range
    })
    print(f"  RMSE: {rmse:.2f} | R²: {r2:.4f} | Out-of-range predictions: {out_of_range}")

# Pick best model (fewest out-of-range first, then lowest RMSE)
results_df = pd.DataFrame(best_models)
print("\n", results_df[["model","rmse","r2","out_of_range"]].sort_values(["out_of_range","rmse"]))

best_idx    = results_df.sort_values(["out_of_range", "rmse"]).index[0]
final_model = best_models[best_idx]["estimator"]
best_name   = best_models[best_idx]["model"]
print(f"\n✅ Best model selected: {best_name}")

# Verify part_time_job effect (must show DIFFERENT scores)
print("\n--- Part-time job effect ---")
s_no  = pd.DataFrame([[5, 75, 7, 7, 0]], columns=features)
s_yes = pd.DataFrame([[5, 75, 7, 7, 1]], columns=features)
print(f"No  part-time: {final_model.predict(s_no)[0]:.1f}%")
print(f"Yes part-time: {final_model.predict(s_yes)[0]:.1f}%")

# Verify mental health effect (must go UP as rating increases)
print("\n--- Mental health effect ---")
for mh in [1, 5, 10]:
    s = pd.DataFrame([[5, 75, mh, 7, 0]], columns=features)
    print(f"  mental_health={mh}: {final_model.predict(s)[0]:.1f}%")

# Save final model
final_model.fit(x, y)
joblib.dump(final_model, "best_model.pkl")
print(f"\n✅ Saved best_model.pkl ({best_name})")
```
